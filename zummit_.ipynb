```python

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
# Make a plot from dictionary
butter_prices = {"Almond butter":10,
                     "Peanut butter": 8,
                     "Cashew butter":12}

fig,ax = plt.subplots()
ax.bar(butter_prices.keys(),butter_prices.values());
ax.set(title= "Nut Butter Store",
      ylabel="Prices ($)");
```

```python
fig, ((ax1, ax2), (ax3, ax4))= plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5))
#Plot to each differnt axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(np.random.randn(1000));

ax1.set(title="Line Graph");
ax2.set(title="Scatter diagram");
ax3.set(title="Bar chat");
ax4.set(title="Histogram");
```
